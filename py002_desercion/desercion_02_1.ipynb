## Data inicial

```python
#Cargar datos
dfi = pd.read_csv("D:/DIFODS/py002_desercion/data/datainicial.csv",sep=";")
df_act= pd.read_csv("D:/DIFODS/py002_desercion/data/carga_actividad.csv",sep=";")
dfi.head()
```

```python
print('Número de registros: '  + str(dfi.shape[0]))
```

```python
#Existen valores duplicados como: USUARIO =604968
dfi=dfi.drop_duplicates() # Eliminar registros duplicados
dfi=dfi[dfi["DNI_CARACTERIZADO"]!=10000000 ] # Eliminar los registros que tienen como DNI_CARACTERIZADO =10000000
print('Número de registros: '  + str(dfi.shape[0]))
```

```python
#Se eliminan los registros que tienen como ninguna actividad en la variable cumplimiento
dfi_1=dfi[dfi['CUMPLIMIENTO']!='NINGUNA ACTIVIDAD']
dfi_1.CUMPLIMIENTO.value_counts()
```

```python
#Comprobar si hay duplicados en los dos grupos de cumplñimiento y al menos una actividad
df_culminaron = dfi_1[dfi_1["CUMPLIMIENTO"]=='CULMINARON']
df_almenos = dfi_1[dfi_1["CUMPLIMIENTO"]=='AL MENOS UNA ACTIVIDAD']
df_culminaron = df_culminaron.iloc[:,[1,3]]
df_almenos = df_almenos.iloc[:,[1,3]]
df_culminaron["DD"] = 1
df_almenos["DO"]= 2
resul=pd.merge(df_culminaron,df_almenos,on=["CURID","DNI_CARACTERIZADO"],how="left")
resul[resul["DO"]>=1]
```

```python
variables = ['CUMPLIMIENTO',
                'CONDICION',
                'DESCRIPCION_SUBTIPO_TRABAJADOR',
                'ESCALA',
                'condicion_laboral',
                'DAREACENSO',
                'NIVEL_NEXUS',
                'SEXO',
                'CAMPUS',
                'CURID',
                'CURSO',
                'USUARIO',
                'act_realizada',
                'Porc_act_realizada',
                'MAX_FECHA',
                'DNI_CARACTERIZADO',
                '%_E1',
                '%_E2',
                '%_E3',
                '%_E4',
                'NOTA_Inducción_CuestionarioEntrada',
                'cursos_matriculados']
```

```python
dfi_2=dfi_1.loc[:,variables]
dfi_2.head()
```

```python
# Eliminar los valor null en las variables de escala , condución_laboral, dareacenso, nivel_nexus
dfi_2=dfi_2.dropna()
dfi_2.isnull().sum()
```

```python
print(dfi_2["ESCALA"].value_counts())
print(" ")
print(dfi_2["condicion_laboral"].value_counts())
# print(" ")
# print(dfi_2["DAREACENSO"].value_counts())
# print(" ")
print(dfi_2["NIVEL_NEXUS"].value_counts())
# print(" ")
# print(dfi_2["SEXO"].value_counts())
```

```python
dfi_2['ESCALA']=pd.to_numeric(dfi_2['ESCALA'], errors='coerce')
dfi_2.condicion_laboral = np.where(dfi_2.condicion_laboral == 'NOMBRADO', "NOMBRADO",np.where(dfi_2.condicion_laboral == 'CONTRATADO',"CONTRATADO",np.where(dfi_2.condicion_laboral == 'DESIGNADO',"DESIGNADO","DESTAQUE_ENCARGADO")))
# dfi_2.DAREACENSO = np.where(dfi_2.DAREACENSO == 'Urbana', 0 ,1 )
# dfi_2.SEXO = np.where(dfi_2.SEXO == "FEMENINO",0,1)
dfi_2.NIVEL_NEXUS = np.where((dfi_2.NIVEL_NEXUS == "Inicial - Jardín") | (dfi_2.NIVEL_NEXUS == "Inicial - Jardín")| (dfi_2.NIVEL_NEXUS == "Inicial - Cuna"),"Inicial",np.where(dfi_2.NIVEL_NEXUS == "Primaria","Primaria",np.where(dfi_2.NIVEL_NEXUS == "Secundaria","Secundaria","BasicaAE")))
dfi_2.ESCALA = np.where(dfi_2.ESCALA == 1,1,np.where(dfi_2.ESCALA == 2,2,np.where(dfi_2.ESCALA == 3,3,np.where(dfi_2.ESCALA == 4,4,np.where(dfi_2.ESCALA == 5,5,np.where(dfi_2.ESCALA == 6,6,np.where(dfi_2.ESCALA == 7,7,8)))))))
```

```python
print(dfi_2["ESCALA"].value_counts())
print(" ")
print(dfi_2["condicion_laboral"].value_counts())
# print(" ")
# print(dfi_2["DAREACENSO"].value_counts())
print(" ")
print(dfi_2["NIVEL_NEXUS"].value_counts())
# print(" ")
# print(dfi_2["SEXO"].value_counts())
```

```python
dicc_cursos={298:28,289:28,280:41}
dfi_2["DURACION"] = dfi_2["CURID"].map(dicc_cursos)
```

```python
# Union de DFi con df_act ( dias de actividad del docente)
dfi_3 = pd.merge(dfi_2, df_act, left_on=["CURID","DNI_CARACTERIZADO"],right_on=["CURID","IDNUMBER"], how="left")
dfi_3["PORC_DIAS"] = round(dfi_3["DIAS"] / dfi_3["DURACION"],2)
dfi_3["Porc_act_realizada"] = round(dfi_3["Porc_act_realizada"] / 100, 2)
#dfi_3 = dfi_3[dfi_3["NOTA_Inducción_CuestionarioEntrada"]>=0]
dfi_3.head()
```

```python
dfi_3.to_csv("D:/datospruebadeser.csv",sep=";")
```

```python
#Seleccionar los datos
x = dfi_3.loc[:,['ESCALA','condicion_laboral','DAREACENSO','NIVEL_NEXUS','SEXO','Porc_act_realizada','cursos_matriculados','%_E1', '%_E2', '%_E3', '%_E4','PORC_DIAS','NOTA_Inducción_CuestionarioEntrada']]
y = dfi_3['CUMPLIMIENTO']

#Clasificacion de datos 
y = np.where(y == "CULMINARON",0,1)
```

```python
x.ESCALA=x.ESCALA.astype('object')
```

```python
x.info()
```

```python
X =pd.get_dummies(x)
X
```

```python
# El modelo 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2 , random_state=123)
print(X_train.shape)
print(X_test.shape)
```

```python
# Configuracion del modelo - Regresion logistica
from sklearn.linear_model import LogisticRegression
regresion_logistica = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=1000,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=3151, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)

#Entrenar el modelo
regresion_logistica.fit(X_train,y_train)
```

```python
#Predecior lo valores 
y_pred = regresion_logistica.predict(X_test)
```

```python
unique, counts = np.unique(y_pred, return_counts=True)
print(unique, counts)
```

```python
#Evaluación de clasificación con el modelo para el curid 24 
from sklearn.metrics import confusion_matrix
matriz_log = confusion_matrix(y_pred, y_test)

# precision_scoreprint modelo anterios
from sklearn.metrics import precision_score
precision = precision_score(y_pred, y_test)

# accuracy_score
from sklearn.metrics import accuracy_score
exactitud = accuracy_score(y_pred, y_test)

# recall_score
from sklearn.metrics import recall_score
sensibilidad = recall_score(y_pred, y_test)
# print('Sensibilidad del modelo:')

# F1 Score
from sklearn.metrics import f1_score
puntajef1 = f1_score(y_pred, y_test)

# ROC - AUC
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_pred, y_test)
gini = roc_auc*2 - 1


print('Matriz Confusion Logistica')
print(matriz_log)
print('Metricas:')
print('')
print('Precisión del modelo:', precision)
print('Exactitud del modelo:', exactitud)
print('Sensibilidad del modelo:', sensibilidad)
print('Puntaje F1 del modelo:', puntajef1)
print('Curva ROC - AUC del modelo:', roc_auc)
print('GINI del modelo:', gini)
```
